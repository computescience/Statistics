In this notebook, I am going to simulate the offer and interview probability using Bayes' Theorem. Given different values of the offer rate (successful), when change the interview probability, what's the final probability getting an offer? This probability question can be solved with Bayes Theorem. These two events are dependent. 
The equation is shown below:


```python
import matplotlib.pyplot as plt
import statistics_practice
import numpy as np
import pandas as pd
```

```python
#construct data
resume = []
interview = []
offer = []
df = pd.DataFrame()

for i in range(10,1010, 10):
    resume.append(i)
for i in np.arange(0.1,1.1, 0.1):
    interview.append(i)
for i in np.arange(0.1,1.1, 0.1):
    offer.append(i)
#calculate with different successful rate of offer, how many resumes submissions.
df['Interview'] = interview
prob = []
```

```python
#Calcuate the probabilities 
for j in offer:
    for k in interview:
        p = statistics_practice.stats.conditionProb(k,j)
        prob.append(p)
    df['OfferProb'+str(round(j,1))] = prob
    prob = []

```

```python
#plot the data
c = df.columns


for c in c:

    plt.plot(df['Interview'],df[c], label = c)
plt.legend(loc='lower right')
plt.xlabel('Interview Probability')
plt.ylabel('Offer Probability')
plt.title("Bayes Theorem on offer probability")
plt.show()
```

Conclusion: for all offer probability, with increase of interview probability, the final offer probability increases. When the offer probability increases to 50%, it becomes the linear correlation. When the offerProba is smaller than 0.5, apparently, the curve becomes power law relation. Chances to get the offer becomes smaller.
